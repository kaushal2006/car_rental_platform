```python
class Car:
    def __init__(
        self,
        car_id,
        car_name,
        car_year,
        number_plate,
        hourly_rate=None,
        daily_rate=None,
        weekly_rate=None,
        is_available_for_rent="N",
        is_rented="N",
    ):
        self.car_id = car_id
        self.car_name = car_name
        self.car_year = car_year
        self.number_plate = number_plate
        self.hourly_rate = hourly_rate
        self.daily_rate = daily_rate
        self.weekly_rate = weekly_rate
        self.is_available_for_rent = is_available_for_rent
        self.is_rented = is_rented

    def __str__(self):
        return f"car_id: {self.car_id}, car_name:{self.car_name}, hourly_rate: {self.hourly_rate}, daily_rate:{self.daily_rate}, weekly_rate:{self.weekly_rate}, is_available_for_rent:{self.is_available_for_rent}, is_rented:{self.is_rented}"

    def setForRental(self, h_rate, d_rate, w_rate, available_for_rent):
        self.hourly_rate = h_rate
        self.daily_rate = d_rate
        self.weekly_rate = w_rate
        self.is_available_for_rent = available_for_rent
        self.is_rented = "N"  # better to reset to N for safety net

    def resetRental(self):
        self.hourly_rate = None
        self.daily_rate = None
        self.weekly_rate = None
        self.is_available_for_rent = "N"
        self.is_rented = "N"
        # mathod(updateIsAvailableForRent)
        # mathod(updateIsAvailableForRent)

    def rentACar(self):
        self.is_rented = "Y"

    def returnACar(self):
        self.is_rented = "N"


# creating list
car_inventory = []

# appending instances to list
car1 = Car(1, "Acura MDX", 2021, "XYZ 1232")
car1.setForRental(10, 100, 500, "Y")
car_inventory.append(car1)

car2 = Car(2, "Toyota Rav4", 2014, "ABC 1324")
car2.setForRental(5, 40, 200, "Y")
car_inventory.append(car2)


car3 = Car(3, "Volvo CS6", 2016, "PQR 1232")
car3.setForRental(8, 60, 250, "Y")
car_inventory.append(car3)

car4 = Car(4, "Toyota Camry", 2022, "OMG 1224")
car4.setForRental(9, 75, 300, "Y")
car_inventory.append(car4)

car5 = Car(5, "Acura RDX", 2019, "MQP 3433")
car5.setForRental(8, 80, 350, "Y")
car_inventory.append(car5)

car6 = Car(6, "BMW X7", 2021, "ABC 3454")
car6.setForRental(10, 110, 800, "Y")
car_inventory.append(car6)

car7 = Car(7, "Hyundai Sonata", 2018, "PQR 1232")
car7.setForRental(9, 60, 240, "Y")
car_inventory.append(car7)

car8 = Car(8, "Tesla Model X", 2023, "OMG 3343")
car8.setForRental(10, 120, 700, "Y")
car_inventory.append(car8)

car9 = Car(
    9, "Lucid Air", 2024, "ELE 1232"
)  # in my example this car is not available for rent
car9.setForRental(30, 200, 1000, "Y")
car_inventory.append(car9)

car10 = Car(10, "Rivion X", 2023, "ELE 3343")
car10.setForRental(15, 190, 800, "Y")
car_inventory.append(car10)

# Accessing object value using a for loop
for car in car_inventory:
    print(car)
```

```python
class Customer:
    def __init__(
        self,
        customer_id,
        customer_name,
        customer_licence_number,
    ):
        self.customer_id = customer_id
        self.customer_name = customer_name
        self.customer_licence_number = customer_licence_number

    def __str__(self):
        return f"customer_id: {self.customer_id}, customer_name:{self.customer_name}, customer_licence_number: {self.customer_licence_number}"


class CustomerRental:
    def __init__(
        self,
        customer_id,
        car_id,
        rent_method,
        rental_start_datetime,
        rental_end_datetime,
        bill,
    ):
        self.customer_id = customer_id
        self.car_id = car_id
        self.rent_method = rent_method
        self.rental_start_datetime = rental_start_datetime
        self.rental_end_datetime = rental_end_datetime
        self.bill = bill

    def __str__(self):
        return f"customer_id: {self.customer_id}, car_id:{self.car_id}, rent_method: {self.rent_method}, rental_start_datetime: {self.rental_start_datetime}, rental_end_datetime: {self.rental_end_datetime}, bill: {self.bill}"
```
